```python
import pandas as pd
VIS= pd.read_excel ('VIS.xlsx')
NO_VIS =pd.read_excel ('NO_VIS.xlsx')
```

```python
Vis =[]
for i in range(79):
    Vis.append(1)
VIS['Vis']=Vis
```

```python
No_Vis =[]
for i in range(40):
    No_Vis.append(0)
NO_VIS['Vis']=No_Vis
```

```python
Proyects= VIS.append(NO_VIS, ignore_index=False, verify_integrity=False, sort=None)
```

```python
Proyects
```

```python
Proyects=Proyects.set_index('Nombre_x')
```

```python
Proyects
```

```python
area = Proyects.groupby('Vis')
area_priv= area['Área_Privada(M2)'].agg('count')
area_priv.to_frame()
```

### Como seria el apto promedio de cada grupo? 

```python
price=[]
area_tot=[]
area_priv=[]
room=[]
bath=[]
nombre=['VIS','NO VIS']
price.append(VIS['Precio(Millones COP)'].agg('sum')/79)
price.append(NO_VIS['Precio(Millones COP)'].agg('sum')/40)
area_tot.append(VIS['Área_Construida(M2)'].agg('sum')/79)
area_tot.append(NO_VIS['Área_Construida(M2)'].agg('sum')/40)
area_priv.append(VIS['Área_Privada(M2)'].agg('sum')/79)
area_priv.append(NO_VIS['Área_Privada(M2)'].agg('sum')/40)
room.append(VIS['Alcobas'].agg('sum')/79)
room.append(NO_VIS['Alcobas'].agg('sum')/40)
bath.append(VIS['Baños'].agg('sum')/79)
bath.append(NO_VIS['Baños'].agg('sum')/40)
mean= pd.DataFrame(list(zip(price,area_tot,area_priv,room,bath,nombre)))
mean.columns = ['Precio', 'Área Total', 'Área Privada','Alcobas','Baños','Vivienda']
mean.set_index('Vivienda')
```
