## Arreglemos esto

```python
import re
geocodes=[]
for i in range(79):
    geocodes.append(VIS['Dirección'][i])

for i in range(79):
    geocodes[i] = geocodes[i].replace('Calle','Cl.')
    geocodes[i] = geocodes[i].replace('No.','#')
    geocodes[i] = geocodes[i].replace('Transversal','TV')
    geocodes[i] = geocodes[i].replace('Diagonal','DG')
    geocodes[i] = geocodes[i].replace('Diagonal','DG')
    geocodes[i] = geocodes[i].replace('Bis','BIS')
    geocodes[i] = geocodes[i].replace('Villa Candelaria','')
    geocodes[i] = geocodes[i].replace(' Altavista Centro Comercial local 1','')   
    geocodes[i] = geocodes[i].replace('esquina','ESQ')
    geocodes[i] = geocodes[i].replace(' piso 1','') 
    geocodes[i] = geocodes[i].replace(' piso 1','')  
for i in range(79):
    geocodes[i] = geocodes[i] + ', Bogotá, Colombia'  
geocodes
```

```python
VIS_MAP_FIX= pd.DataFrame(list(zip(Nombre,geocodes)))
VIS_MAP_FIX.columns = ['Nombre','Geocodigo'] 
VIS_MAP_FIX =VIS_MAP_FIX.set_index('Nombre')
VIS_MAP_FIX
```

```python
from geopy.extra.rate_limiter import RateLimiter
from geopy.distance import geodesic as dist
locator = Nominatim(user_agent='myGeocoder')
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
VIS_MAP_FIX['location'] = VIS_MAP_FIX['Geocodigo'].apply(geocode)
VIS_MAP_FIX['point'] = VIS_MAP_FIX['location'].apply(lambda loc: tuple(loc.point) if loc else None)
VIS_MAP_FIX[['latitude', 'longitude', 'altitude']] = pd.DataFrame(VIS_MAP_FIX['point'].tolist(), index=VIS_MAP_FIX.index)
```

```python
mapa2= VIS_MAP_FIX.dropna()
len(mapa2)
```

```python
from geopy.extra.rate_limiter import RateLimiter
import folium
import numpy as np
import folium.plugins as plugins
folium_map = folium.Map(location=[4.60971,-74.08175],
                        zoom_start=12,
                        tiles='CartoDB dark_matter')


plugins.FastMarkerCluster(data=list(zip(mapa2['latitude'].values, mapa2['longitude'].values))).add_to(folium_map)
folium.LayerControl().add_to(folium_map)
folium_map
```
